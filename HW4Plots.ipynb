```python
qlResults = pd.read_csv('results/ql/results.csv',header=None)
qlResults.columns = ['iteration','steps','time','lr']
# fig = plt.subplot()
num_iters = 300
for x in np.linspace(0.1, .6, 6):
    print x
    fig = qlResults[qlResults.lr ==x].groupby('iteration').mean()['steps'].loc[:num_iters].plot(kind = 'line',label='Learning Rate-'+str(x))
fig.set_xlabel('Number of Iteration')
fig.set_ylabel('Number of Steps')
fig.set_title('QLearning Iteration Convergence')
fig.set_xticks(np.arange(0,num_iters,25))
fig.set_xlim(100)
fig.set_ylim(0,100)
fig.legend()
fig.grid(True)
```
